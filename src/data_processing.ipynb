```python
def get_top_level_directories_and_files(path, sort_func, file_extension):
    dirs, files = [], []
    for d in os.listdir(path):
        if os.path.isdir(os.path.join(path, d)):
            dirs.append(d)
        else:
            if d.endswith(file_extension):
                files.append(d)
    return sorted(dirs, key=sort_func), files

dirs, files = get_top_level_directories_and_files(DATASET_PATH, lambda x: x.split(' ')[-1], '.json')
print(f"SubDir: {dirs}")
print(f"Files: {files}")
```

We can see that images in different `Tiles` directories vary in size. Therefore, we'll resize the images to a single uniform size, which will be based on our pre-defined patch size. Deep learning models often process images in batches of fixed-size inputs. For uniform size patches, we can

- vectorize input processing
- avoid dynamic shape handling
- save on memory and runtime efficiency

CNNs for satellite image tasks (e.g., UNet, ResNet) often require inputs of fixed shape.

If patches include padding or trimmed areas

- models might learn edge padding as a feature
- predictions could be biased or noisy near borders

In land cover classification or semantic segmentation, uniform patches ensure all labelled regions are processed equally.

#### Processing Pipeline

1. Read images and masks for each tile.
2. Transform BGR to RGB (since cv2 reads images in BGR order).
3. Define patch size and patchify both images and masks for uniform size.
4. Normalize images (not masks) such that their value is between 0 and 1.

Let's write some functions for the processing pipeline.
  

```python
PATCH_SIZE = 256
```

```python
def get_adjusted_shape_for_image(shape):    
    if len(shape) == 3:
        return [(s // PATCH_SIZE)*PATCH_SIZE for s in shape[:-1]] + [shape[-1]]
    return [(s // PATCH_SIZE)*PATCH_SIZE for s in shape]
```

```python
print(get_adjusted_shape_for_image((658, 682, 3)))
```

```python
processed_images, processed_masks = [], []
for d in subdir:
    BASE_IMAGE_PATH = os.path.join(path, d, 'images')
    BASE_MASK_PATH = os.path.join(path, d, 'masks')
    for f in os.listdir(BASE_IMAGE_PATH):
        name = os.path.splitext(f)[0]
        image = cv2.imread(os.path.join(BASE_IMAGE_PATH, f))
        mask = cv2.imread(os.path.join(BASE_MASK_PATH, name + '.png'))
        # since the class labels for the masks give RGB code, we'll convert the masks to RGB
        # we can also apply the conversion to images (BGR2RGB), but it's optional (if we are training a model from scratch)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
        # Normalizing images so that their value remains between 0 and 1. 
        image = image / 255.0
        # Patchify both images and masks for uniform sizes
        image_patches = patchify(image, (PATCH_SIZE, PATCH_SIZE, 3), step=PATCH_SIZE)
        mask_patches = patchify(mask, (PATCH_SIZE, PATCH_SIZE, 3), step=PATCH_SIZE)
        # image_patches are usually (N, 1, h, w, c)
        i, j = image_patches.shape[:2]
        # print(i,j)
        for ii in range(i):
            for jj in range(j):
                processed_images.append(np.squeeze(image_patches[ii, jj, 0, :, :, :]))
                processed_masks.append(np.squeeze(mask_patches[ii, jj, 0, :, :, :]))
processed_images = np.array(processed_images)
processed_masks = np.array(processed_masks)

TOTAL_IMAGES = len(processed_images)
TOTAL_MASKS = len(processed_masks)
print(f"Total images and masks: ({TOTAL_IMAGES}, {TOTAL_MASKS})")   
```

Now we need to work on the labels. From Kaggle website, we find the following classes and their hex labels

1. Building: #3C1098
2. Land (unpaved area): #8429F6
3. Road: #6EC1E4
4. Vegetation: #FEDD3A
5. Water: #E2A929
6. Unlabeled: #9B9B9B

Now RGB represents Red, Green, Blue where each channel has values between 0 and 255. In the hex code, we can decode (say) `#3C1098` as

- `3C` -> `3 x 16 + 12 x 1` -> `60`
- `10` -> `16`
- `98` -> `9 x 16 + 8` -> `152`

```python
def hex_to_rgb(hex_code):
    # hex code starts with #
    return tuple(int(hex_code[i:i+2], 16) for i in [1, 3, 5])

def rgb_to_hex(r, g, b):
    return (f'#{r:02X}{g:02X}{b:02X}')

print(hex_to_rgb('#3C1098'))
print(rgb_to_hex(60, 16, 152))
```

```python
class_label_str = [
    'Building: #3C1098', 
    'Land (unpaved area): #8429F6', 
    'Road: #6EC1E4', 
    'Vegetation: #FEDD3A', 
    'Water: #E2A929', 
    'Unlabeled: #9B9B9B']
class_labels_rgb = {}
for s in class_label_str:
    k, v = s.split(': ')
    class_labels_rgb.update({k: hex_to_rgb(v)})

NUM_CLASSES = len(class_labels_rgb)
print(NUM_CLASSES)
class_labels_rgb
```

Now based on the above, generate a label dataset from the masks

```python
label_map = {k: idx for idx, k in enumerate(class_labels_rgb.keys())}
label_map
```

Now we want to place these labels instead of RGB values in the mask data

```python
rgb_to_label_map = {class_labels_rgb.get(k): label_map.get(k) for k in label_map.keys()}
rgb_to_label_map
```

```python
def rgb_mask_to_label(mask, mapping):
    h, w, _ = mask.shape
    label_mask = np.zeros((h, w), dtype=np.uint8)

    for rgb, label in mapping.items():
        matches = np.all(mask == rgb, axis=-1)
        label_mask[matches] = label

    return label_mask
```

```python
labels = []
for mask in processed_masks:
    labels.append(rgb_mask_to_label(mask, rgb_to_label_map))
```

```python
np.array(labels).shape
```

```python
labels_categorical = to_categorical(labels, num_classes=NUM_CLASSES)
```

```python
X_train, X_test, y_train, y_test = train_test_split(processed_images, labels_categorical, test_size=0.05, random_state=51)
```
